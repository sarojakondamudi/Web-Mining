```python
# 2 .Plot MSE versus order m for m = 1, 2, 3, 4, 5, 6, 7, 8. Identify the best choice of order m

import matplotlib.pyplot as plt 
plt.style.use('seaborn-whitegrid') 
import numpy as np 
noise_scale = 100  
number_of_samples = 50
x =  25*(np.random.rand(number_of_samples, 1) - 0.8) 
y = 5 * x + 20 * x**2 + 1 * x**3 + noise_scale*np.random.randn(number_of_samples, 1) 
#plt.plot(x,y,'ro') 
X1=x.flatten()
Y1=y.flatten()
Z1=np.polyfit(X1,Y1,1)  
Z2=np.polyfit(X1,Y1,2)  
Z3=np.polyfit(X1,Y1,3)  
Z4=np.polyfit(X1,Y1,4)  
Z5=np.polyfit(X1,Y1,5)  
Z6=np.polyfit(X1,Y1,6)  
Z7=np.polyfit(X1,Y1,7)  
Z8=np.polyfit(X1,Y1,8) 
mse1 = np.square(np.subtract(np.polyval(Z1,x),y).mean())
mse2 = np.square(np.subtract(np.polyval(Z2,x),y).mean())
mse3 = np.square(np.subtract(np.polyval(Z3,x),y).mean())
mse4 = np.square(np.subtract(np.polyval(Z4,x),y).mean())
mse5 = np.square(np.subtract(np.polyval(Z5,x),y).mean())
mse6 = np.square(np.subtract(np.polyval(Z6,x),y).mean())
mse7 = np.square(np.subtract(np.polyval(Z7,x),y).mean())
mse8 = np.square(np.subtract(np.polyval(Z8,x),y).mean())
x_ax=[1,2,3,4,5,6,7,8]
y_ax=[mse1,mse2,mse3,mse4,mse5,mse6,mse7,mse8]
plt.plot(x_ax,y_ax,'g--')
```

```python
# From the above its clearly visible that m = 3 gives the best fit and hence i'll be considering m=3 going on
# the m kept changing from 4 and at 6 was very high so ot is safe to take m = 3
```
