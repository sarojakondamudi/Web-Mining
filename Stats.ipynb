```python
# Part 5: Construct a confidence interval with your data
import scipy.stats as stats
import math

sample_10_wr = group_data_wr_100[4]
sample_mean_10_wr = statistics.mean(sample_10_wr)
z_critical_10_wr = stats.norm.ppf(q = 0.975)  # Get the z-critical value check about this
print("z-critical value:")              # Check the z-critical value
print(z_critical_10_wr)                        
pop_stdev_10_wr = statistics.stdev(raw_data['Close_ETF'])
print(pop_stdev_10_wr) # Get the population standard deviation
margin_of_error_10_wr = z_critical_10_wr * (pop_stdev_10_wr/math.sqrt(len(sample_10_wr)))
confidence_interval_10_wr = (sample_mean_10_wr - margin_of_error_10_wr,
                       sample_mean_10_wr + margin_of_error_10_wr)  
print("Confidence interval for 10 random samples is ")
print(confidence_interval_10_wr)
```
